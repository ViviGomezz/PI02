```python
import pandas as pd
import numpy as np
```

```python
localidades_acceso= pd.read_csv('/content/drive/MyDrive/PI02/Listadodelocalidadesconconectividadainternet_2750221689223165308.csv')
```

```python
localidades_acceso.replace('--', 0, inplace=True)
localidades_acceso.replace('SI', 1, inplace=True)
```

```python
Columnas_si= ['ADSL','CABLEMODEM', 'DIALUP',	'FIBRAOPTICA', 'WIRELESS', 'SATELITAL']
localidades_acceso['Acceso_internet'] = localidades_acceso[Columnas_si].eq(1).any(axis=1).map({True: '1'})
localidades_acceso.fillna('0', inplace=True)
```

```python
localidades_acceso.to_csv('Listadodelocalidadesconconectividadainternet', index=False)
```
